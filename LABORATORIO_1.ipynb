```python
# Cajero Automático
import random #Importar liberia random para generar saldo aleatorio

# Generar un saldo inicial aleatorio entre 100 y 5000
saldo = random.randint(100, 5000)
intentos = 3  # Número de intentos permitidos para ingresar el PIN

# Solicitar y almacenar el PIN del usuario
pin_correcto = input("Ingrese su PIN de seguridad: ")  # Guardamos el PIN como cadena de texto

# Bucle - Solicitar PIN con intentos limitados
while intentos > 0:
    pin = input("Ingrese su PIN: ")  # Solicitar el PIN al usuario

    if pin == pin_correcto:  # Comparación correcta entre cadenas
        print("Acceso concedido")

        # Menú de opciones dentro de un bucle
        while True:
            print("\n1. Consultar saldo")
            print("2. Depositar dinero")
            print("3. Retirar dinero")
            print("4. Transferir dinero")
            print("5. Salir")
            opcion = input("Seleccione una opción: ")

            if opcion == "1":
              #Consultar saldo
                print(f"Saldo actual: ${saldo}")

            elif opcion == "2":
              #Depositar dinero
                cantidad = float(input("Ingrese la cantidad a depositar: "))
                if cantidad > 0:
                    saldo += cantidad #Se suma depósito al saldo actual
                    print(f"Depósito exitoso. Nuevo saldo: ${saldo}")
                else:
                    print("La cantidad a depositar debe ser mayor a 0.")

            elif opcion == "3":
              #Retirar Dinero de la cuenta
                cantidad = float(input("Ingrese la cantidad a retirar: "))
                if 0 < cantidad <= saldo:
                    saldo -= cantidad # Se resta el retiro del saldo actual
                    print(f"Retiro exitoso. Nuevo saldo: ${saldo}")
                else:
                    print("Fondos insuficientes o cantidad no válida.")

            elif opcion == "4":
              #Transferencia de dinero a otra cuenta
                cantidad = float(input("Ingrese la cantidad a transferir: "))
                destinatario = input("Ingrese el destinatario: ")
                if 0 < cantidad <= saldo:
                    saldo -= cantidad # Se resta el monto transferido del saldo actual
                    print(f"Transferencia de ${cantidad} realizada a {destinatario}. Nuevo saldo: ${saldo}")
                else:
                    print("Fondos insuficientes o cantidad no válida.")

            elif opcion == "5":
              #Salir del cajero
                print("Gracias por usar el cajero automático. Hasta luego.")
                break  # Rompe el bucle del menú y finaliza el programa

            else:
                print("Opción no válida. Intente de nuevo.")

        break  # Sale del bucle de autenticación
    else:
        intentos -= 1
        print(f"PIN incorrecto. Intentos restantes: {intentos}")

# Si el usuario agota los intentos, la cuenta se bloquea
if intentos == 0:
    print("Cuenta bloqueada. Demasiados intentos incorrectos.")
```
