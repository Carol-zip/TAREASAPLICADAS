```python
#Punto 1
import numpy as np
import matplotlib.pyplot as plt
import random

def fibonacci(n):
    a, b = 0, 1 # A va a tener el valor de 0 y B el valor de 1 ya que la secuencia Fibonacci empieza con esos 2 numeros
    fib_sequence = [] #esto de aquí es una lista vacía para que vaya guardando los datos y pueda graficar correctamente

    for i in range(n):
        fib_sequence.append(a)
        c = a + b
        a = b
        b = c

    return fib_sequence

# Generar los primeros 20 números de Fibonacci
n = 20
fib_numbers = fibonacci(n)

# Graficar la sucesión
plt.plot(range(n), fib_numbers, marker="o", linestyle="-", color="b", label="Fibonacci")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.title("Sucesión de Fibonacci")
plt.legend()
plt.grid()
plt.show()
```

```python
#Punto 2
import numpy as np
import matplotlib.pyplot as plt
import random

def fibonacci(n):
    fib_series = [0, 1]
    for i in range(2, n):
        fib_series.append(fib_series[-1] + fib_series[-2])
    return fib_series[:n]

def plot_fibonacci(n):
    series = fibonacci(n)
    plt.plot(series, marker='o', linestyle='-', color='b')
    plt.title(f'Serie de Fibonacci (n={n})')
    plt.xlabel('Índice')
    plt.ylabel('Valor')
    plt.grid()
    plt.show()

def analyze_temperatures(temperatures):
    mean_temp = np.mean(temperatures)
    median_temp = np.median(temperatures)
    std_dev_temp = np.std(temperatures)
    print(f"Media: {mean_temp:.2f}, Mediana: {median_temp:.2f}, Desviación estándar: {std_dev_temp:.2f}")

    months = np.arange(1, len(temperatures) + 1)  # Corregida la indentación
    plt.plot(months, temperatures, marker='o', linestyle='-', color='r')
    plt.title('Temperatura Media Mensual')
    plt.xlabel('Meses')
    plt.ylabel('Temperatura (°C)')
    plt.grid()
    plt.show()

def simulate_dice_rolls():
    rolls = [random.randint(1, 6) for _ in range(1000)]
    values, counts = np.unique(rolls, return_counts=True)
    plt.bar(values, counts, color='g', alpha=0.7)
    plt.xticks(values)
    plt.title('Frecuencia de lanzamientos de un dado')
    plt.xlabel('Número del dado')
    plt.ylabel('Frecuencia')
    plt.show()

# Ejecución de las funciones
temperatures = [25.6, 26.1, 24.8, 27.3, 28.4, 26.7, 25.9, 29.1, 30.2, 27.8, 24.3, 23.9, 26.5, 28.0, 27.4, 29.3, 30.5, 26.8, 25.7, 24.6]

n = int(input("Ingrese el número de términos de la serie de Fibonacci: "))
plot_fibonacci(n)

analyze_temperatures(temperatures)

simulate_dice_rolls()
```

```python
#Punto 4
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores eje x
x = np.linspace(0, 2*np.pi, 100)  # 100 valores de x entre 0 y 2π

# Funciones trigonométricas
y_sin = np.sin(x)   # Seno de x
y_cos = np.cos(x)   # Coseno de x
y_tan = np.tan(x)   # Tangente de x

# Configurar etiquetas
plt.figure(figsize=(8,6))  # Tamaño del gráfico
plt.title("Funciones Trigonométricas")
plt.xlabel("Eje X") # Nombre eje x
plt.ylabel("Eje Y") # Nombre eje y

# Graficar las funciones
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.plot(x, y_tan, label='tan(x)', linestyle='dashed', color='green')

# Ajustes adicionales
plt.axhline(0, color='black', linewidth=0.5)  # Línea horizontal en y=0
plt.axvline(0, color='black', linewidth=0.5)  # Línea vertical en x=0
plt.legend()  # Mostrar la leyenda de las funciones
plt.grid()  # Agregar cuadrícula
plt.show()  # Mostrar el gráfico
```

```python
#Punto 5
import matplotlib.pyplot as plt
import numpy as np

# Definir los valores de x
x = np.linspace(0, 10, 100)  # 100 valores entre 0 y 10

# Funciones
y_sqrt = np.sqrt(x) # y = raiz cuadrada de x
y_x = x  # y = x

plt.figure(figsize=(8,6))
plt.plot(x, y_sqrt, label=r'$y=\sqrt{x}$', color='blue') # Gráfica de raíz cuadrada
plt.plot(x, y_x, label=r'$y=x$', color='red', linestyle='dashed') # Línea y = x

# Configurar el gráfico
plt.title("GráficaPunto 5")  # Título
plt.xlabel("Eje X")  # Nombre del eje X
plt.ylabel("Eje Y")  # Nombre del eje Y
plt.legend()  # Mostrar la leyenda
plt.grid()  # Agregar cuadrícula

# 5. Mostrar el gráfico
plt.show()
```

```python
#Punto 6
import matplotlib.pyplot as plt
import numpy as np

# Definir los puntos
x1, y1 = 2, 3  # Punto 1 (x1, y1)
x2, y2 = 6, 7  # Punto 2 (x2, y2)

# Calcular la pendiente
m = (y2 - y1) / (x2 - x1)  # Fórmula de la pendiente

# Generar valores de x para la recta
x = np.linspace(x1, x2, 100)  # 100 valores entre x1 y x2
y = m * (x - x1) + y1  # Ecuación de la recta

# Graficar los puntos y la recta
plt.plot(x, y, label="Recta", color="blue")  # Dibuja la recta
plt.scatter([x1, x2], [y1, y2], color="red", label="Puntos")  # Dibuja los puntos

# Configurar el gráfico
plt.title("Gráfica de una recta entre dos puntos")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.legend()
plt.grid()

# Mostrar el gráfico
plt.show()
```
