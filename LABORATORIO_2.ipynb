Como se muestra en la grafica, hay 3 puntos de corte, los cuales no son numeros enteros y su factorizazion ser√≠a mjuy complicada, para ello usamos el metodo de nweton pars aproximar las raices de la funciom, vamos a sacar los intervalos de losp untos de corte

-1,0
2,3

```python
def newton_raphson(f, fprime, x0, tol, maxiter):
    x = x0
    for i in range(maxiter):
        fx = f(x)
        if abs(fx) < tol:
            return x

        fp = fprime(x)
        if fp == 0:
            print('Error la derivada es cero')
            break
        x = x - fx / fp
    return x

# Ejemplo de uso:
f=lambda x: x**5 - 6*x**3 + 2*x -2
g=lambda x: 5*x**4 - 18*x**2 + 2
x0 = -3
tol = 1e-5
maxiter = 10000
root = newton_raphson(f, g, x0, tol, maxiter)
print(f"La ra√≠z de la funci√≥n es: {root}")
```
